```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
PATH = '../datasets/'
```

```python
def clean_dataset(df):
    # Transpose the dataframe and set the first row as the header
    df_t = df.T
    new_header = df_t.iloc[0] # setting first row as header
    df_t = df_t[1:] # take the data less the header row
    df_t.columns = new_header # set the header row as the df header
    
    # Strip leading and trailing spaces from the column names
    df_t.columns = df_t.columns.str.strip()
    
    # Fill missing values with 0
    df_t.fillna(value=0, inplace=True)
    
    # Ensure correct data types
    for column in df_t.columns:
        # If the column can be converted to numeric, convert it
        df_t[column] = pd.to_numeric(df_t[column], errors='ignore')
    
    # Remove duplicate rows
    df_t.drop_duplicates(inplace=True)
    
    # Remove duplicate columns (if applicable)
    df_t = df_t.loc[:,~df_t.columns.duplicated()]
    
    return df_t
```

```python
def load(file_path, sheet_name):
    return pd.read_excel(file_path, sheet_name)
```

```python
def clean_quarterly_dataset(df):
    # Transpose the dataframe and set the first row as the header
    df_t = df.T
    new_header = df_t.iloc[0]  # set the first row as header
    df_t = df_t[1:]  # take the data less the header row
    df_t.columns = new_header  # set the header row as the df header
    df_t.columns = df_t.columns.str.strip()

    df_t = df_t.apply(pd.to_numeric, errors='ignore')
    df_t.fillna(value=0, inplace=True)
    df_t.drop_duplicates(inplace=True)
    
    df_t = df_t.loc[:, ~df_t.columns.duplicated()]
    
    return df_t
```
